```python
import findspark
findspark.init()
from pyspark.sql import SparkSession
from pyspark.sql.types import StructType, StructField, StringType, IntegerType
spark = SparkSession.builder.appName("example").getOrCreate()
```

```python
data = [
    {"Columna 1 Big Data": "Spark", "Columna 2 Big Data": "PYTHON"},
    {"Columna 1 Big Data": "Spark", "Columna 2 Big Data": "PYTHON"},
    {"Columna 1 Big Data": "Spark", "Columna 2 Big Data": "PYTHON"}
]

# Define the schema (optional, but recommended for performance)
schema = StructType([
    StructField("Columna 1 Big Data", StringType(), True),
    StructField("Columna 2 Big Data", StringType(), True)
])

# Create a DataFrame
df = spark.createDataFrame(data, schema=schema)

# Show the DataFrame
df.show()
```
